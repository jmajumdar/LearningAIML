```python
# function to plot a boxplot and a histogram along the same scale.


def histogram_boxplot(data, feature, figsize=(12, 7), kde=False, bins=None):
    """
    Boxplot and histogram combined

    data: dataframe
    feature: dataframe column
    figsize: size of figure (default (12,7))
    kde: whether to the show density curve (default False)
    bins: number of bins for histogram (default None)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,  # Number of rows of the subplot grid= 2
        sharex=True,  # x-axis will be shared among all subplots
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize,
    )  # creating the 2 subplots
    sns.boxplot(
        data=data, x=feature, ax=ax_box2, showmeans=True, color="violet"
    )  # boxplot will be created and a star will indicate the mean value of the column
    sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins, palette="winter"
    ) if bins else sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2
    )  # For histogram
    ax_hist2.axvline(
        data[feature].mean(), color="green", linestyle="--"
    )  # Add mean to the histogram
    ax_hist2.axvline(
        data[feature].median(), color="black", linestyle="-"
    )  # Add median to the histogram
```

```python
# function to create labeled barplots


def labeled_barplot(data, feature, perc=False, n=None):
    """
    Barplot with percentage at the top

    data: dataframe
    feature: dataframe column
    perc: whether to display percentages instead of count (default is False)
    n: displays the top n category levels (default is None, i.e., display all levels)
    """

    total = len(data[feature])  # length of the column
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count + 1, 5))
    else:
        plt.figure(figsize=(n + 1, 5))

    plt.xticks(rotation=90, fontsize=15)
    ax = sns.countplot(
        data=data,
        x=feature,
        palette="Paired",
        order=data[feature].value_counts().index[:n].sort_values(),
    )

    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(
                100 * p.get_height() / total
            )  # percentage of each class of the category
        else:
            label = p.get_height()  # count of each level of the category

        x = p.get_x() + p.get_width() / 2  # width of the plot
        y = p.get_height()  # height of the plot

        ax.annotate(
            label,
            (x, y),
            ha="center",
            va="center",
            size=12,
            xytext=(0, 5),
            textcoords="offset points",
        )  # annotate the percentage

    plt.show()  # show the plot
```

## **K-Means Clustering**

Before we can do any clustering, we must scale the data. As we have seen, the data for credit limits has a wide range of 197,000 compared to the data for number of credit cards or number of visits to bank, which range in 10 to 15. This will make the clustering heavily influenced by the credit limits.

```python
# We need to make a copy of the data before scaling so that we preserve the non-scaled data for use later.
mycopy = df.copy()
```

```python
# I will use the standard scaler to do a z-score scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(mycopy)
```

```python
# The scaler doesn't give me a dataframe, so I have to make it into a dataframe, and I need the original column names for the DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=mycopy.columns)
```

### **Use elbow method to try to figure the value of K for the K-Means clustering**

The way I will do it is using a technqiue we learned in the MLS session, where we loop through different values of K and plot the graph to get the elbow plot.

The code here is adapted from what we did in MLS session.

```python
clusters = range(1, 9)
meanDistortions = []

for k in clusters:
    model = KMeans(n_clusters=k)
    model.fit(scaled_df)
    prediction = model.predict(scaled_df)
    distortion = (
        sum(
            np.min(cdist(scaled_df, model.cluster_centers_, "euclidean"), axis=1)
        ) / scaled_df.shape[0]
    )
    meanDistortions.append(distortion)

plt.plot(clusters, meanDistortions, "bx-")
plt.xlabel("k")
plt.ylabel("Average Distortion")
plt.title("Selecting k with the Elbow Method", fontsize=20)
```

### **Observation**

The sharp turn in the curve (the elbow) appears to be for K = 3. The elbow method suggests that the optimal value of K is **K=3** for the K-Means clustering.

But let me check with the sihouette score as well.

### **Silhouette score**

```python
# This code is adapted from the one we used in MLS class
sil_score = []
cluster_list = list(range(2, 10))
for n_clusters in cluster_list:
    clusterer = KMeans(n_clusters=n_clusters)
    preds = clusterer.fit_predict((scaled_df))
    # centers = clusterer.cluster_centers_
    score = silhouette_score(scaled_df, preds)
    sil_score.append(score)

plt.plot(cluster_list, sil_score)
```

### **Observation**

I see that the silhouette score is highest for k=3, so comparing the silhouette score for many clusters also gives us the same resule of optimal K is **K=3**.

I also see that the silhouette score is only a little bit above 0.5 and nowhere close to 1. This indicates that while this may be the best clustering possible, it is not strongly separated from other clusters.

I expect that my boxplot analysis of the various clusters will show an overlap.

### **Decision for K**
The results from the elbow technique and the silhouette scores are in agreement, so I will use a value of 3 for the K-Means clustering.

```python
# Create the 3 clusters using KMeans clustering.
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(scaled_df)
```

```python
# The kmeans.labels_ identifies the various labels (clusters) that data has been assigned to.
# But I need to associate this data with the original (unscaled) data so that I can actually see the original values and the clusters they were assigned to.
# Let me assign a label that tells me that this is the label obtained from K-Means clustering so that I can later compare results from the hierarchical
# clustering as well
df["K_means_segments"] = kmeans.labels_
```

## **Hierarchical Clustering**

We had previously scaled the data, so we do not need to spend any time rescaling the data. We will use the `scaled_data` variable to do the hierarchical clustering.




```python
# Instead of trying to do the hierarchical clustering using a variety of ways to 
# measure distance, I am going to stick the the EUCLIDEAN distance.
# However, I will try a variety of linkage methods to find the highest cophenetic
# distance for the various clusters.

# list of linkage methods
linkage_methods = ["single", "complete", "average", "centroid", "ward", "weighted"]

high_cophenet_corr = 0
high_dm_lm = [0, 0]

for lm in linkage_methods:
    Z = linkage(scaled_df, metric="euclidean", method=lm)
    c, coph_dists = cophenet(Z, pdist(scaled_df))
    print("Cophenetic correlation for {} linkage is {}.".format(lm, c))
    if high_cophenet_corr < c:
        high_cophenet_corr = c
        high_dm_lm[0] = "euclidean"
        high_dm_lm[1] = lm
```

```python
# I have my highest cophenetic correlation
print(
    "Highest cophenetic correlation is {}, which is obtained with {} linkage.".format(
        high_cophenet_corr, high_dm_lm[1]
    )
)
```

**Using Euclidean distance measure, I got the highest cophenetic correlation score using Average Linkage.**

Now I want to display the dendograms for the various linkages used to figure out the appropriate number of clusters. It is possible that the diagrams show me a better distribution using a different linkage method even though the cophenetic score is lower.


```python
# I already have the list of linkage methods in the code exxecuted above, so I
# don't need to define it again.

# lists to save results of cophenetic correlation calculation
compare_cols = ["Linkage", "Cophenetic Coefficient"]

# to create a subplot image
fig, axs = plt.subplots(len(linkage_methods), 1, figsize=(15, 30))

# We will enumerate through the list of linkage methods above
# For each linkage method, we will plot the dendrogram and calculate the cophenetic correlation
for i, method in enumerate(linkage_methods):
    Z = linkage(scaled_df, metric="euclidean", method=method)

    dendrogram(Z, ax=axs[i])
    axs[i].set_title(f"Dendrogram ({method.capitalize()} Linkage)")

    coph_corr, coph_dist = cophenet(Z, pdist(scaled_df))
    axs[i].annotate(
        f"Cophenetic\nCorrelation\n{coph_corr:0.2f}",
        (0.80, 0.80),
        xycoords="axes fraction",
    )
```

### **Observations**
* The dendogram with single linkage looks extremely messy with no clear hierarchical separation. It also has a low cophenetic correlation score of 0.74, so I will reject this.
* The dendogram for the Complete Linkage appears to be a lot cleaner in terms of separation of cluster, and it has a cophenetic correlation score of 0.86.
* The dendogram for the Average Linkage is not very clean and there isn't good separation of clusters, even though the cophenetic correlation score is 0.90. I will reject this.
* The dendogram for the Centroid Linkage is not very clean and there isn't good separation of clusters, even though the cophenetic correlation score is 0.89. I will reject this.
* The dendgram for the Ward Linkage has done a good job of separating out the clusters, even though the cophenetic correlation score is only 0.74. In fact, the distances between the clusters appears to be the highest for the Ward linkage.
* The dendogram for the Weighted Linkage has doen a poor job of separating the clusters, even though it has a cophenetic correlation score of 0.89.

### **Clustering Decision**
* I am going to proceed with the Ward linkage method for the hierarchical clustering.
* The appropriate value of K appears to be 3 from looking at the dendogram.

```python
# Let's do a clustering using ward linkage, euclidean distance, and 3 clusters.
HCmodel = AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage="ward")
HCmodel.fit(scaled_df)
```

```python
# Assign the labels back to the original data. Use a separate label to not override the classification we had got from K-Means clustering.
df["HC_segments"] = HCmodel.labels_
```
